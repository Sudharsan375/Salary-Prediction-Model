#Salary Prediction


*   Problem Statment Mr.Premchand is an HR in Sprout Knowledge Solutions. He needs to predict the salary based on the candidate's experience, test score, and interview score.Based on the given data set create your own machince learning model to predict the salary.


# 1.Understanding the business problem and requirement

* Mr.Premchand wants to predict the salary of a candidate based on their attributes and based on the Machine Learning model that uses the data set.

#2.Data Collection

```python
import pandas as pd
df=pd.read_csv('/content/input_dataset.csv')
df
```

#3.Data Processing

* Handling the missing data

```python
df.isnull()
```

```python
#handling the missing data
import math
test_score_median = math.floor(df['test_score'].median())
print(test_score_median)
```

```python
df['test_score'] = df['test_score'].fillna(test_score_median)
df['experience'] = df['experience'].fillna('zero')   #filling the empty experience with zero
df
```

```python
#installing word to number libary
!pip install word2number
```

```python
from word2number import w2n
```

```python
df['experience'] = df['experience'].apply(lambda x:w2n.word_to_num(x))
df
```

# 4.Data Visualization

```python
import matplotlib.pyplot as plt
x = df['experience']
y = df['salary']
plt.scatter(x, y)
plt.show()
```

#5.Segregate the Independent and Depedendent variables

```python
X = df.drop('salary',axis = 1)
X
```

```python
y = df['salary']
y
```

#6.Building **ML** model 

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[4,7,8]])
```

```python
reg.score(X,y)
```

```python
#our model is 93% accurate which is a good score and we can use our model. 
```

#7.Downloading the model

* We can use the ML model on other project which make our process efficient where we need not to do data collection and processing on the data set.

```python
import pickle
with open ('candidates_hiring_model','wb') as f:
  pickle.dump(reg,f)
```
